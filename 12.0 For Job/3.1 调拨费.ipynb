# 2.1 数据框1：调拨费明细

```python
df1_usecols_keep=["创建时间","调拨完成时间","调拨单号","商家编码","规格名称","源仓库","目标仓库","调拨数量","备注"]
df1_usecols_rename=["申请日期","完成日期","调拨单号","商家编码","规格名称","调出仓库","调入仓库","调拨数量","调拨原因"]

df1_usecols_keep_to_rename = dict(zip(df1_usecols_keep,df1_usecols_rename))
df1_usecols_keep_to_rename
```

# 2.2 数据框2：单品列表(去重)

```python
df2 = pd.read_csv(input_file2)
```

**tips:确保商家编码唯一值，`避免笛卡尔集`**

```python
df2.duplicated(subset=["商家编码"]).value_counts()
```

```python
df2.drop_duplicates(subset=["商家编码"],inplace=True,ignore_index=True)

df2 = df2.reindex(columns=["商家编码","品牌","毛重","体积","每箱数量","重/泡"],fill_value=0)

df2["重/泡"] = df2["重/泡"].map({"泡":"泡货","重":"重货"})
```

```python
df2.duplicated(subset=["商家编码"]).value_counts()
```

# 3.1 df1调拨明细 和 df2单品列表 合并(merge)

```python
df1[:5]
```

```python
df2[:5]
```

```python
df1.columns
```

```python
df2.columns
```

```python
df1_left_merge_df2 = pd.merge(df1,df2,how="left",on="商家编码")

df1_left_merge_df2
```

```python
df1_left_merge_df2["调出仓库区域"] = ""
df1_left_merge_df2["调入仓库区域"] = ""
df1_left_merge_df2["箱数"] = df1_left_merge_df2["调拨数量"]/df1_left_merge_df2["每箱数量"]
df1_left_merge_df2["体积"] = df1_left_merge_df2["体积"]/1000000
df1_left_merge_df2["总体积 (m³)"] = df1_left_merge_df2["体积"]*df1_left_merge_df2["调拨数量"]
df1_left_merge_df2["毛重"] = df1_left_merge_df2["毛重"]/1000
df1_left_merge_df2["总重量 (TO)"] = df1_left_merge_df2["毛重"]*df1_left_merge_df2["调拨数量"]
```

```python
df1_left_merge_df2.columns
```

```python
df1_left_merge_df2 = df1_left_merge_df2.rename(columns={"每箱数量":"箱规","重/泡":"货品类型","毛重":"单位重量 (TO)","体积":"单位体积 (m³)"})
```

```python
df1_left_merge_df2_reindex = ["申请日期","完成日期","调拨单号","商家编码","品牌","规格名称","调出仓库","调出仓库区域",
 "调入仓库","调入仓库区域","调拨数量","箱规","箱数","调拨原因","货品类型","单位体积 (m³)",
 "总体积 (m³)","单位重量 (TO)","总重量 (TO)"]
```

```python
df1_left_merge_df2 = df1_left_merge_df2.reindex(columns=df1_left_merge_df2_reindex)
```
