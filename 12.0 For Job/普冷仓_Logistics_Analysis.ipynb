def Tier(weight):
    if weight / 1000 > 10 :
        Tier = "X >10kg"
    elif weight / 1000 > 5 :
        Tier = "X >5kg,≤10kg"
    elif weight / 1000 > 3 :
        Tier = "X >3kg,≤5kg"
    elif weight / 1000 > 2 :
        Tier = "X >2kg, ≤3kg"
    elif weight / 1000 > 1 :
        Tier = "X >1kg, ≤2kg"
    elif weight / 1000 > 0.5 :
        Tier = "X >0.5kg, ≤1kg"
    elif weight / 1000 > 0.3 :
        Tier = "X >0.3kg, ≤0.5kg"
    else :

    return Tier

#### 1.4 Pivot

````python
df_pivot = df.pivot_table(index=["月份","店铺/货主","仓库区域","省份","重量段 (快递费)","物流公司"],
                          values=["物流单号/运单号","订单量","计费重量","总重量 (kg)","快递费","货品数量/拣货数量"],
                          aggfunc={"物流单号/运单号":"count",'订单量':"sum",
                                   "计费重量":"sum","总重量 (kg)":"sum","快递费":"sum",
                                  '货品数量/拣货数量':"sum"})

df_pivot
````

````python
df_pivot.rename(columns={"物流单号/运单号":"包裹量"},inplace=True)
````

````python
df_pivot = df_pivot.reset_index()

df_pivot
````
